# Import the necessary libraries

```python
import detecto
import matplotlib.pyplot as plt
from torchvision import transforms
from detecto.utils import read_image
from detecto.core import Model, Dataset, DataLoader
```

# Load the dataset

```python

path = r'C:\Users\MMb\Desktop\detecto/'
sampled_image = '51l4PUDGfcL.jpg'
image = read_image(path + sampled_image)
plt.imshow(image)
plt.show()

dataset = Dataset(path)
image, targets = dataset[4]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

# Define custom transforms

```python

custom_transforms = transforms.Compose([
    transforms.ToPILImage(),
    transforms.Resize(800),
    transforms.RandomHorizontalFlip(0.5),
    transforms.ColorJitter(saturation=0.2),
    transforms.ToTensor(),
])
```

# Create the dataset with custom transforms

```python

dataset = Dataset(path, transform=custom_transforms)
```

# Define the classes

```python

classes = ['non_veg', 'veg']
```

# Define the model

```python

model = Model(classes)
```

# Train the model

```python

loader = DataLoader(dataset, batch_size=2, shuffle=True)
model.fit(loader, verbose=True)
```

# Perform inference

```python

val_dataset = Dataset(r'C:\Users\MMb\Desktop\netscribes\assg_12\data_eng_assn\validation_detecto_dataset/')
losses = model.fit(dataset, val_dataset, epochs=15, learning_rate=0.01,
                   gamma=0.2, lr_step_size=5, verbose=True)

plt.plot(losses)
plt.show()
```
