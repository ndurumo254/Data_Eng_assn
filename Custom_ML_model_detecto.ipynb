# Custom ML model to detect Veg / Non-Veg Symbol from the Image.

```python
import matplotlib.pyplot as plt
from torchvision import transforms
from detecto.utils import read_image
from detecto.core import Model, Dataset, DataLoader
from detecto.visualize import show_labeled_image
```

```python
path= r'C:\Users\MMb\Desktop\detecto/'
```

```python
sampled_image='51l4PUDGfcL.jpg'
```

```python
image= read_image(path+sampled_image)
plt.imshow(image)
plt.show()
```

```python
dataset = Dataset(path)
```

```python
image, targets = dataset[4]
show_labeled_image(image, targets['boxes'], targets['labels'])
```

```python
classes=['non_veg','veg']
```

```python
model = Model(classes)
```

```python
losses = model.fit( dataset, epochs=15,
                   learning_rate=0.001, verbose=True)
```

```python
image= read_image(path+"61lMO5wUicL._SX679_.jpg")
labels, box, scores =model.predict(image)
print('labels:', labels)
print('Box:', box)
print('scores:', scores)
```

```python
show_labeled_image(image, box[0], labels[0])
```
